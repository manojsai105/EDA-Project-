```python
Hardware_engineer_jobs = ['Technical Support Engineer', 'Network Administrator', 'System Engineer', 'Systems Engineer']
Programmer_analyst_jobs = ['programmer analyst' , 'programmer analyst trainee']
Software_Engineer_jobs = ['Java Software Engineer', 'Programmer Analyst', 'Systems Analyst', 'Software Engineer'
'Java Developer', 'Assistant Software Engineer', 'Software Developer', 'Research Associate'
'Senior Software Engineer', 'Trainee Software Engineer', 'System Engineer', '.NET Developer'
'Project Engineer', 'Senior Java Developer', 'Software Development Engineer', 'Associate Software Engineer'

'Software Trainee', 'Assistant Systems Engineer', '.NET Web Developer', 'Assistant System Engineer'
'Software Engineer Associate', 'Software Analyst', 'Junior Software Engineer', 'Software Executive'
'Graduate Engineer Trainee' , 'Technology Analyst', 'Software Engineering Associate']
Associate_Engineer_jobs = ['Assistant Software Engineer', 'Associate Software Engineer' ,'Assistant Systems Engineer']

Hardware_engineer_jobs = [job.lower() for job in Hardware_engineer_jobs]
Programmer_analyst_jobs = [job.lower() for job in Programmer_analyst_jobs]
Software_Engineer_jobs = [job.lower() for job in Software_Engineer_jobs]
Associate_Engineer_jobs = [job.lower() for job in Associate_Engineer_jobs]
```

```python
hardware_engineer_data = df[df['Designation'].isin(Hardware_engineer_jobs)]
programmer_analyst_data = df[df['Designation'].isin(Programmer_analyst_jobs)]
software_engineer_data = df[df['Designation'].isin(Software_Engineer_jobs)]
associate_engineer_data = df[df['Designation'].isin(Associate_Engineer_jobs)]
```

```python
hardware_engineer_data = df[df['Designation'].isin(Hardware_engineer_jobs)]
programmer_analyst_data = df[df['Designation'].isin(Programmer_analyst_jobs)]
software_engineer_data = df[df['Designation'].isin(Software_Engineer_jobs)]
associate_engineer_data = df[df['Designation'].isin(Associate_Engineer_jobs)]
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(11, 11))
axes = axes.flatten()
sns.boxplot(x='Designation', y='Salary', data=hardware_engineer_data, ax=axes[0])
axes[0].set_title('Hardware Engineer')
axes[0].tick_params(axis='x', rotation=90)
sns.boxplot(x='Designation', y='Salary', data=programmer_analyst_data, ax=axes[1])
axes[1].set_title('Programming Analyst')
axes[1].tick_params(axis='x', rotation=90)
sns.boxplot(x='Designation', y='Salary', data=software_engineer_data, ax=axes[2])
axes[2].set_title('Software Engineer')
axes[2].tick_params(axis='x', rotation=90)
sns.boxplot(x='Designation', y='Salary', data=associate_engineer_data, ax=axes[3])
axes[3].set_title('Associate Engineer')
axes[3].tick_params(axis='x', rotation=90)
plt.tight_layout()
plt.show()
```

```python
avg_salary_hardware_engineer = hardware_engineer_data["Salary"].mean()
avg_salary_programmer_analyst = programmer_analyst_data["Salary"].mean()
avg_salary_software_engineer = software_engineer_data["Salary"].mean()
avg_salary_associate_engineer = associate_engineer_data["Salary"].mean()
avg_salaries = [avg_salary_hardware_engineer, avg_salary_programmer_analyst, avg_salary_software_engineer, avg_salary_associate_engineer]
designations = ['Hardware Engineer', 'Programming Analyst', 'Software Engineer', 'Associate Engineer']
plt.figure(figsize=(10, 6))
plt.bar(designations, avg_salaries)
plt.title('Avg Salary by Designation')
plt.xlabel('Designation')
plt.ylabel('Avg Salary')
plt.show()
```

```python
all_jobs = Hardware_engineer_jobs + Programmer_analyst_jobs + Software_Engineer_jobs + Associate_Engineer_jobs
avg_salary_by_designation = {}
for job in all_jobs:
    job_data = df[df['Designation'].str.lower() == job]
    avg_salary = job_data['Salary'].mean()
    avg_salary_by_designation[job] = avg_salary
avg_salary_series = pd.Series(avg_salary_by_designation)
# Create a bar plot
plt.figure(figsize=(14, 8))
ax = avg_salary_series.sort_values().plot(kind='bar', color='skyblue')
plt.title('Average Salary by Designation')
plt.xlabel('Designation')
plt.ylabel('Average Salary')
plt.xticks(rotation=90)
# Draw a border line at 250,000
ax.axhline(y=250000, color='red', linestyle='--', linewidth=2, label='Border Line: ₹250,000')
plt.legend()
plt.show()
```

```python
from scipy.stats import f_oneway

programming_analyst_salary = programmer_analyst_data["Salary"]
software_engineer_salary = software_engineer_data["Salary"]
hardware_engineer_salary = hardware_engineer_data["Salary"]
associate_engineer_salary = associate_engineer_data["Salary"]
f_statistic, p_value = f_oneway(programming_analyst_salary,software_engineer_salary,hardware_engineer_salary,associate_engineer_salary)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There are significant differences in average salaries.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in average salaries.")
```

Relationship between Gender and specialization

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Specialization', hue='Gender', data=df)
plt.title('Specialization vs Gender')
plt.xlabel('Specialization')
plt.ylabel('Count')
plt.xticks(rotation=90, ha='right') # Rotate x-axis labels for better readability
plt.show()
```

```python
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(df['Gender'], df['Specialization'])
print("Contingency Table:")
print(contingency_table)
chi2, p, _, _ = chi2_contingency(contingency_table)
print("\nChi-squared statistic:", chi2)
print("P-value:", p)
alpha = 0.05 # Set the significance level
print("\nSignificance level:", alpha)
if p < alpha:
    print("There is a significant relationship between Gender and Specialization.")
else:
    print("There is no significant relationship between Gender and Specialization.")
```
